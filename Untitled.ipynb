## Machine Learning

```python
def euclidean_dist(point1,point2):
    import math
    squared_dist=0
    for i  in range(len(point1)):
        squared_dist+=math.pow(point1[i]-point2[i],2)
    return math.sqrt(squared_dist)
```

```python
def mode(labels):
    from collections import Counter
    return Counter(labels).most_common(1)[0][0]
```

```python
def knn(data,query,k):
    points_distance=[]
    for index,example in enumerate(data):
        distance=euclidean_dist(example[:-1],query)
        points_distance.append((distance,index))
    sorted_points_distance=sorted(points_distance)
    K_sorted_points=sorted_points_distance[:k]
    k_nearest_labels=[data[i][1] for distance,i in K_sorted_points]
    return mode(k_nearest_labels)
```

```python
clf_data = [
       [22, 1],
       [23, 1],
       [21, 1],
       [18, 1],
       [19, 1],
       [25, 0],
       [27, 0],
       [29, 0],
       [31, 0],
       [45, 0],
    ]
query=[20]
pred=knn(clf_data,query,3)
print('predicted class is {}'.format(pred))
```
